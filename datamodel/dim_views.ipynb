```
Replace view adlste_coa.coa_dim_site
as locking row for access
Select
 LogDate
,Sales_Region as Region
,Sales_Geography as Geo
,Account_Name
,Site_ID
,Site_Category
,System_Use
,substr(Teradata_SW_Version,1,5) as DBS_Version
,case 
 when All_Platforms is null then
     case when Site_Category like '%Amazon%' then 'AWS'
          when Site_Category like '%Azure%'  then 'Azure'
          when Site_Category like '%TDVM%'   then 'VMWare'
          when Site_Category like '%TDMana%' then 'TDCloud'
          else 'unknown' end 
 when All_Platforms = '7011' then 'IFX2.1'
 when All_Platforms = '7010' then 'IFX2.0'
 when All_Platforms = '7000' then 'IFX1.0'
 when All_Platforms = '7200' then 'IBase1'
 when All_Platforms = '7210' then 'IBase2'
 when All_Platforms = '200CL' then 'CCU'
 when All_Platforms like '1800%' then '1800'
 when All_Platforms like '1700%' then '1700'
 when All_Platforms like '1750%' then '1750'
 else All_Platforms end as Platforms
,TotalNumTPANodes as Active_Nodes
,TotalNumHSNNodes as HotStandby_Nodes
,Machine_Learning_Engine
from adlste_coa.coa_dim_cis_siteID2
where Teradata_Site = 'Y'
  and Active_Nodes >0
```
